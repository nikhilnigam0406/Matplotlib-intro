```python
from matplotlib import style
```

```python
classes = ["History","Civil","Maths","Physice","DSA","C/Java"]
dataset_01 = [74,57,67,65,45,95]
dataset_02 = [45,78,54,93,47,58]
```

```python
plt.bar(classes, dataset_01)
```

```python
plt.barh(classes, dataset_01,color = "g")
```

```python
# Methods
plt.bar(classes, dataset_01,width = 0.1, align = "edge",color="m", linewidth = 505,alpha  = 0.9,
        linestyle = "--",label = "class 1 students")
```

```python
style.use("ggplot")
plt.figure(figsize = (11,9))
plt.bar(classes, dataset_01,width = 0.3, align = "edge",color="m", linewidth = 15,alpha  = 0.6,
        linestyle = "--",label = "class 1 students")
plt.title("Subjects", fontsize = "8")

plt.bar(classes, dataset_02,width = 0.1, align = "edge",color="m", linewidth = 505,alpha  = 0.9,
        linestyle = "--",label = "class 1 students")
```
