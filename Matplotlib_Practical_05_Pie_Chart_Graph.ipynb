```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Creation
classes =["DSA","AI-ML","C","JAVA","ENG"]
class1_students = [45,75,45,25,96]
plt.pie([1])
plt.show()
```

```python
# simple class
classes =["DSA","AI-ML","C","JAVA","ENG"]
class1_students = [45,75,45,25,96]

plt.pie(class1_students, labels = classes)
plt.show()
```

```python
classes =["DSA","AI-ML","C","JAVA","ENG"]
class1_students = [45,75,45,25,96]
# slocing
explode = [0,0.03,0,.1,0.025]
# colors
colors = ["y","b","g",'black',"r"]
#textprop
textprope = {"fontsize":10}
plt.pie(class1_students,
        labels = classes,
        explode = explode,
        colors = colors,
        autopct= "%.2f%%",
        shadow = True,
        startangle=270,
       textprops = textprope,
       center=(0, 38),
       radius = 0.7)
plt.legend()

plt.show()
```
